```python
%time
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import solve
import random
import copy
import time
import numpy as np

def is_edge(point1,point2):
     global point
    # print(point1)
     #print(point2)
     #竖直线
     if point2[0]==point1[0]:
             flag_all_large=0
             flag_all_less=0
             for i in point:
                if (i[0]-point1[0])>=0:
                    flag_all_large=flag_all_large+1
                if(i[0]-point1[0])<=0:
                    flag_all_less=flag_all_less+1
             if flag_all_large==len(point) or flag_all_less==len(point):
                 plt.plot([point1[0],point2[0]],[point1[1],point2[1]],color='r')
                 plt.scatter(point1[0],point1[1],color='g')
                 plt.scatter(point2[0],point2[1],color='g')
    #一般线(包括水平线)
     else:
             k = (point2[1]-point1[1])/(point2[0]-point1[0])
             b =  point2[1]-k*point2[0]
             flag_all_large=0
             flag_all_less=0
             for i in point:
                if (k*i[0]+b-i[1])>=0:
                    flag_all_large=flag_all_large+1
                if(k*i[0]+b-i[1])<=0:
                    flag_all_less=flag_all_less+1
             
             if flag_all_large==len(point) or flag_all_less==len(point):
                 plt.scatter(point1[0],point1[1],color='y',marker=(9, 2, 30))
                
                
             
    
x=[[1,2],[2,5]]
y=[[3,4],[5,6]]
point=[]
for i in range(100):
    point.append([random.randint(0,100),random.randint(0,100)])
temp_x=[]
temp_y=[]
for i in point:
    temp_x.append(i[0])
    temp_y.append(i[1])
plt.scatter(temp_x,temp_y,marker=(9, 1, 30))
for i in range(len(point)):
    for j in range(i+1,len(point)):     
        is_edge(point[i],point[j])
print("ok")
```
